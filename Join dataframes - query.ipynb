```
%sql
SELECT *
FROM ejemplo
JOIN ejemplo2
  ON ejemplo.id==ejemplo2.id
WHERE valor RLIKE '(\\d)'
```

```
%sql
WITH contadorDeAltoYBajo AS (SELECT altobajo, COUNT(*) as contador
FROM tablaconaltosybajos 
GROUP BY altobajo)
SELECT *
FROM tablaconaltosybajos as t
JOIN contadorDeAltoYBajo as c
    ON c.altobajo = t.altobajo
WHERE contador > 5
```
