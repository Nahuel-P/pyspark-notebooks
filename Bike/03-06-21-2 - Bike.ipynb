```
path_station_data = "/databricks-datasets/definitive-guide/data/bike-data/201508_station_data.csv"
path_trip_data = "/databricks-datasets/definitive-guide/data/bike-data/201508_trip_data.csv"
```

```
df_station = spark.read.csv(path_station_data, header= True)
df_trip = spark.read.csv(path_trip_data, header=True)
df_station.show(truncate=False)
df_trip.show(5, False)
```

```
df_trip.createOrReplaceTempView("trips")
df_station.createOrReplaceTempView("station")
```

```
spark.sql("set spark.sql.legacy.timeParserPolicy=LEGACY")
```

```
from pyspark.sql.functions import year, month, dayofmonth,hour, to_date, col

df_trip=df_trip.withColumn("nuevaFecha", to_date(col("Start Date"), "mm/dd/yyyy") )

df = df_trip \
  .select(
    "Trip ID", "nuevaFecha", 
    year(df_trip["nuevaFecha"]).alias('year'), 
    )

display(df)
df.printSchema()
df.select("year").dtypes
```

```
from pyspark.sql.types import StringType
df_yearString = df.withColumn("year",df["year"].cast(StringType()))
df_yearString.select("year").dtypes
```

```
df_yearDecimal=df_yearString.withColumn("yearNumber", df_yearString["year"].cast("decimal(6,2)"))
df_yearDecimal.show()
```
