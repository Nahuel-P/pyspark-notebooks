### Cantidad de minutos entre 2 fechas

```
%sql
SELECT 
  `Trip ID`, 
  Duration,
  `Start Date`,
  `End Date`,
  round((unix_timestamp(`End Date`, "MM/dd/yyyy HH:mm") - unix_timestamp(`Start Date`, "MM/dd/yyyy HH:mm"))/60, 2) AS DurationInMinutes
FROM trip_filtrado
```

### Tiempo de viaje promedio por ciudad

```
%sql
WITH TablaConPromedios AS (SELECT 
  `Trip ID`, 
  Duration,
  `Start Station`,
  `Start Date`,
  `End Date`,
  round((unix_timestamp(`End Date`, "MM/dd/yyyy HH:mm") - unix_timestamp(`Start Date`, "MM/dd/yyyy HH:mm"))/60, 2) AS DurationInMinutes
FROM trips)
SELECT `Start Station`, round(avg(DurationInMinutes)) as promedioDeviajeEnMinutos
FROM TablaConPromedios
GROUP BY `Start Station`
```

### Extraer la fecha y separarlo en 3 columnas

```
from pyspark.sql.functions import year, month, dayofmonth,hour, to_date, col

df_trip=df_trip.withColumn("nuevaFecha", to_date(col("Start Date"), "mm/dd/yyyy") )

df = df_trip \
  .select(
    "Trip ID", "nuevaFecha", 
    year(df_trip["nuevaFecha"]).alias('year'), 
    month(df_trip["nuevaFecha"]).alias('month'), 
    dayofmonth(df_trip["nuevaFecha"]).alias('day')
  #,hour(df_trip["nuevaFecha"]).alias('hour')
    )

display(df)
```
